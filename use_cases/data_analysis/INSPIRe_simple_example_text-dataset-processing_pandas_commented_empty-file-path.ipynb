#### Here's the code after running submitting the previous instructions to the LLM:

```python
import pandas as pd

# Function to determine the file type and appropriately load the data
def load_data(file_path):
    if file_path.endswith('.xlsx'):
        return pd.read_excel(file_path)
    elif file_path.endswith('.csv'):
        return pd.read_csv(file_path)
    else:
        raise ValueError("Unsupported file format. Please provide an Excel or CSV file.")

# Function to process the file and add the 'Sentiment' column
def process_file(file_path):
    # Load the data with the appropriate function based on file type
    df = load_data(file_path)
    
    # Check if the 'Product Rating' column exists
    if 'Product Rating' not in df.columns:
        raise KeyError("'Product Rating' column not found in the provided file.")
    
    # Add the 'Sentiment' column based on 'Product Rating'
    df['Sentiment'] = df['Product Rating'].apply(lambda x: 'Positive' if x >= 4 else 'Negative')
    
    return df

# Example usage
# Replace 'path_to_your_file.xlsx' or 'path_to_your_file.csv' with the actual file path
file_path =  "your_file_path_here.xlsx"
processed_df = process_file(file_path)

# Display a sample of rows taken of the processed DataFrame
print(processed_df.sample(5))
```

### Brand Analysis
**Objective**: Understand the market share and customer perception of each brand. <br>
**Method**: Perform a similar group-by operation as above on the "Brand" column and combine it with sentiment and rating analysis.<br>
**Reasoning**: This helps identify which brands are more popular, how they are rated, and the general sentiment towards them.

```python
# Import necessary libraries for EDA
import matplotlib.pyplot as plt
import seaborn as sns

# Set the visualisation style
sns.set(style="whitegrid")

# 1. Count the number of reviews per brand
brand_review_count = processed_df['Brand'].value_counts()

# 2. Calculate the average product rating for each brand
brand_rating_average = processed_df.groupby('Brand')['Product Rating'].mean()

# 3. Analyze the sentiment distribution for each brand
brand_sentiment_distribution = processed_df.groupby('Brand')['Sentiment'].value_counts(normalize=True).unstack().fillna(0)

# Display the results
fig, axes = plt.subplots(3, 1, figsize=(10, 15))

# Plot the number of reviews per brand
sns.barplot(x=brand_review_count.index, y=brand_review_count.values, ax=axes[0])
axes[0].set_title("Number of Reviews per Brand")
axes[0].set_xlabel("Brand")
axes[0].set_ylabel("Number of Reviews")

# Plot the average product rating for each brand
sns.barplot(x=brand_rating_average.index, y=brand_rating_average.values, ax=axes[1])
axes[1].set_title("Average Product Rating per Brand")
axes[1].set_xlabel("Brand")
axes[1].set_ylabel("Average Rating")

# Plot the sentiment distribution for each brand
brand_sentiment_distribution.plot(kind='bar', stacked=True, ax=axes[2])
axes[2].set_title("Sentiment Distribution per Brand")
axes[2].set_xlabel("Brand")
axes[2].set_ylabel("Percentage of Sentiments")

plt.tight_layout()
plt.show()
```

#### Analyze which brands received the most 4 and 5-star reviews, we need to follow these steps:

**Filter the reviews**: Isolate reviews where the "Product Rating" is either 4 or 5.<br>
**Count the number of 4 and 5-star reviews for each brand**: This gives us an idea of how many high ratings each brand has received.<br>
Implement this analysis to see the distribution of high ratings across different brands.

```python
# 1. Filter the reviews for 4 and 5-star ratings
high_rating_df = processed_df[processed_df['Product Rating'] >= 4]

# 2. Count the number of 4 and 5-star reviews for each brand
high_rating_brand_count = high_rating_df['Brand'].value_counts()

# Plot the results
plt.figure(figsize=(10, 6))
sns.barplot(x=high_rating_brand_count.index, y=high_rating_brand_count.values)
plt.title("Number of 4 and 5-Star Reviews per Brand")
plt.xlabel("Brand")
plt.ylabel("Number of 4 and 5-Star Reviews")
plt.xticks(rotation=45)
plt.show()
```

#### Analyze the distribution of positive versus negative sentiment for the top 10 most reviewed products, we need to follow these steps:

**Identify the top 10 most reviewed products**: Based on the previous frequency analysis. <br>
**Filter the dataset to include only these top 10 products**.<br>
**Calculate the sentiment distribution for each of these products**: Focusing on positive and negative sentiments.<br>
Implement this analysis to understand the sentiment distribution for the top 10 most reviewed products.

```python
# 1. Identify the top 10 most reviewed products
product_review_count = processed_df['Product Reviewed'].value_counts().sort_values(ascending=False)
top_10_products = product_review_count.head(10).index

# 2. Filter the dataset to include only these top 10 products
top_10_data = processed_df[processed_df['Product Reviewed'].isin(top_10_products)]

# 3. Calculate the sentiment distribution for each of these products
top_10_sentiment_distribution = top_10_data.groupby('Product Reviewed')['Sentiment'].value_counts(normalize=True).unstack().fillna(0)

# Plot the sentiment distribution
top_10_sentiment_distribution.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title("Sentiment Distribution for Top 10 Reviewed Products")
plt.xlabel("Product Reviewed")
plt.ylabel("Percentage of Sentiments")
plt.legend(title="Sentiment")
plt.show()
```

#### Perform text analysis on the "Comment" column and extract common themes or features mentioned in the reviews, we will use a word cloud, as it is an effective visual tool for identifying the most frequent words. Here's how we can do it:

**Combine all comments**: Concatenate all the text from the "Comment" column to have a single large text string.<br>
**Text preprocessing**: Remove common stopwords (like "the", "a", "in") which don't add significant meaning to the text analysis.<br>
**Generate the word cloud**: Create a visual representation of the most common words in the comments.<br>
Let's proceed with this analysis:

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

# 1. Combine all comments into a single text string
all_comments = ' '.join(processed_df['Comment'].dropna())

# 2. Text preprocessing - define stopwords to exclude from the analysis
stopwords = set(STOPWORDS)

# 3. Generate the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white', stopwords=stopwords, max_words=100).generate(all_comments)

# Plot the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # Hide the axes
plt.show()
```
